```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('data.csv',',',error_bad_lines=False)
```

```python
data.head()
```

```python
data[data['password'].isnull()]
```

```python
data.dropna(inplace=True)
```

```python
passwords_tuple=np.array(data)
```

```python
passwords_tuple
```

```python
import random
random.shuffle(passwords_tuple)
y=[d[1]for d in passwords_tuple]
X=[d[0]for d in passwords_tuple]
```

```python
X
```

```python
y
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.countplot(x='strength',data=data)
```

```python
def word_divide(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide)
X=vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
feature_names=vectorizer.get_feature_names()
#tfidf vector the first document
first_document_vector=X[0]
#printing the scores
df=pd.DataFrame(first_document_vector.T.todense(),index=feature_names,columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

#  Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
log_classifier=LogisticRegression(penalty='l2',multi_class='ovr',verbose=1)
log_classifier.fit(X_train,y_train)
```

```python
print(log_classifier.score(X_test,y_test))
```

```python
log_classifier=LogisticRegression(random_state=0,multi_class='multinomial',solver='newton-cg',verbose=1)
log_classifier.fit(X_train,y_train)
print(log_classifier.score(X_test,y_test))
```

```python
x_predict=np.array(["@@@@@$%%&7^$#@@@@12345KunalHelloworld"])
x_predict=vectorizer.transform(x_predict)
y_predict=log_classifier.predict(x_predict)
print(y_predict)
```
